### Missing Data
- Lets check for missing data and think if dropping this data is a good idea or not:

```python
#First we get the null values of the dataframe and the percentage of null values:
total = train.isnull().sum().sort_values(ascending=False)
percentage = (train.isnull().sum()/train.isnull().count()).sort_values(ascending=False)*100
```

```python
#Now we can create a dataframe with this two columns:
df_missing = pd.DataFrame(columns = ["Total Null Values","%"])
df_missing["Total Null Values"] = total
df_missing["%"] = percentage
df_missing.head(20)
```

- In the kaggle notebook there is the standard suggestion that if more than 15% of data is missing then we should consider dropping this variables (maybe some investigation regarding this topic is a good idea but since the notebook is well valued then lets trust him)
- We can see that after electrical there are is no missing data so lets consider this variables and get an idea if we can drop all the columns that have missing data even though the missing data might be lower than 15%
    1. Garage Variables: We can consider dropping all this variables since the garage variables that have high correlation with the data are already being taken into account
    2. The same argument can be given with the Bsmt variables.


- Lets drop the columns with more than 1 missing data since we have seen that they are not so important.
- Since we have only 1 missing data from electrical we will drop this observation.
- When dropping this columns we have to drop the columns of test too

```python
train = train.drop((df_missing[df_missing['Total Null Values']>1]).index, 1)
test = test.drop((df_missing[df_missing['Total Null Values']>1]).index, 1)  
train = train.drop((train.loc[train['Electrical'].isnull()]).index)
```

```python
print(train.isnull().sum().max())
```

- Now before tackling the prediction we need to change categorical variables into numerical variables
- Firstly we will split the column in categorical and numerical and then assign the numerical values to the categorical frame
- Finally we can concatenate the variables:

```python
numerical = train._get_numeric_data()
categorical = train
categorical = categorical.drop(numerical.columns,1)
```

Lets check the dataframes:

```python
df_train
```

I will sum this up in a function since it can be a useful function to have. The purpose of this function is to enter a frame with mixed categorical and numerical data and return a dataframe with only numerical values (categorical values have a number assigned now)

```python
def categoricalToNum(df):
    numerical = df._get_numeric_data()
    categorical = df
    categorical = categorical.drop(numerical.columns,1)
    dummies = pd.get_dummies(categorical)
    final_df = pd.concat([dummies,numerical],axis = 1)
    return final_df
```

We can now test this function with the train and test datasets:

```python
df_train = categoricalToNum(train)
df_test = categoricalToNum(test)
```
